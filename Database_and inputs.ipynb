```python
import mysql.connector

# Connect to MySQL server
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="ssm"   # change this
)
cursor = conn.cursor()

# Create database
cursor.execute("CREATE DATABASE IF NOT EXISTS hospital")
print("Database created!")

conn.close()
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="ssm",
    database="hospital"
)
cursor = conn.cursor()

# Create tables
cursor.execute("""
CREATE TABLE IF NOT EXISTS Patients (
    patient_id INT PRIMARY KEY AUTO_INCREMENT,
    name VARCHAR(100) NOT NULL,
    age INT,
    gender ENUM('Male','Female','Other'),
    contact_info VARCHAR(150),
    address TEXT,
    emergency_contact VARCHAR(150)
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS HealthRecords (
    record_id INT PRIMARY KEY AUTO_INCREMENT,
    patient_id INT,
    blood_pressure_systolic INT,
    blood_pressure_diastolic INT,
    heart_rate INT,
    blood_sugar_level DECIMAL(5,2),
    weight DECIMAL(5,2),
    height DECIMAL(5,2),
    BMI DECIMAL(5,2),
    date_recorded DATE,
    bp_normal BOOLEAN,
    hr_normal BOOLEAN,
    sugar_normal BOOLEAN,
    FOREIGN KEY (patient_id) REFERENCES Patients(patient_id)
)
""")


print("Tables created!")

conn.close()
from faker import Faker
import random
import mysql.connector
from datetime import datetime, timedelta

fake = Faker()

# Connect to database
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="ssm",
    database="hospital"
)
cursor = conn.cursor()

# Insert 50 patients
for _ in range(50):
    cursor.execute("""
        INSERT INTO Patients (name, age, gender, contact_info, address, emergency_contact)
        VALUES (%s, %s, %s, %s, %s, %s)
    """, (
        fake.name(),
        random.randint(18, 80),
        random.choice(["Male", "Female"]),
        fake.phone_number(),
        fake.address(),
        fake.phone_number()
    ))

conn.commit()

# Get patient IDs
cursor.execute("SELECT patient_id FROM Patients")
patient_ids = [row[0] for row in cursor.fetchall()]

# Insert 14 days of health records per patient
for pid in patient_ids:
    start_date = datetime.today() - timedelta(days=14)
    for i in range(14):
        date = start_date + timedelta(days=i)
        sys = random.randint(90, 160)
        dia = random.randint(60, 100)
        hr = random.randint(50, 120)
        sugar = round(random.uniform(70, 200), 2)
        weight = round(random.uniform(50, 90), 2)
        height = round(random.uniform(1.5, 1.9), 2)
        bmi = round(weight / (height**2), 2)

        # Normal checks
        bp_normal = 90 <= sys <= 120 and 60 <= dia <= 80
        hr_normal = 60 <= hr <= 100
        sugar_normal = 70 <= sugar <= 140

        cursor.execute("""
            INSERT INTO HealthRecords (
                patient_id, blood_pressure_systolic, blood_pressure_diastolic,
                heart_rate, blood_sugar_level, weight, height, BMI, date_recorded,
                bp_normal, hr_normal, sugar_normal
            ) VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
        """, (pid, sys, dia, hr, sugar, weight, height, bmi, date.date(),
              bp_normal, hr_normal, sugar_normal))

conn.commit()
print("Inserted 2 weeks records for 50 patients!")

conn.close()

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="ssm",
    database="hospital"
)
cursor = conn.cursor()

cursor.execute("""
SELECT p.name, AVG(h.heart_rate)
FROM Patients p
JOIN HealthRecords h ON p.patient_id = h.patient_id
GROUP BY p.patient_id
ORDER BY p.patient_id DESC
LIMIT 5
""")

for row in cursor.fetchall():
    print(row)

conn.close()
```
